```python
import numpy as np

#Parameters
r = np.array([100, 50, 10, 1], dtype=float)
p = np.array([1/3, 1/12, 1/4, 1/3], dtype=float)
c = 0.5  # channel capacity

N = len(r)
#here we find the ratio of each class where 2nd class has the highest ratio as explained in the notebook
ratio = r / p

# Step 1: sort classes by decreasing ratio
#here as we know the ratio now, the lets put them in order [1.0.2.3]
order = np.argsort(-ratio)

# fractional knapsack
u = np.zeros(N)             # admission probabilities [1 - accept , 0 - reject]
remaining_capacity = c

for j in order:
    if p[j] <= remaining_capacity: # does the class fits in remaining capacity
        u[j] = 1.0 #if yes then put it in
        remaining_capacity -= p[j]
    else: #if not
        u[j] = remaining_capacity / p[j] #admit a fraction
        remaining_capacity = 0.0 # tell that no more capacity left
        break

# ----- Results -----
print("Class | Reward | Probability | Ratio | Admit prob")
for j in range(N):
    print(f"{j+1:5d} | {r[j]:6.1f} | {p[j]:11.4f} | {ratio[j]:6.1f} | {u[j]:.3f}")
```
